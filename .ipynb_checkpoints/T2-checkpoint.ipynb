```python
import os
os.environ["OMP_NUM_THREADS"] = "1" 
os.environ["MKL_NUM_THREADS"] = "1" 

from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
customer_data = pd.read_csv('Wholesale customers data.csv')
```

```python
customer_data.describe()
```

```python
plt.scatter(customer_data.iloc[:,3].values, customer_data.iloc[:,4].values)
plt.xlabel("Grocerys", size=12)
plt.ylabel("Milk", size=12)


plt.ylim(0, 20000)
plt.xlim(0, 40000)

plt.show()
```

```python
# 2. Boxplots
# Fr att f en verblick av knsfrdelningen i datan. 
fig, axes = plt.subplots(1, 3, figsize=(25, 5))
sns.boxplot(customer_data, x="Channel", y="Grocery", ax = axes[0])
sns.boxplot(customer_data, x="Channel", y="Milk" ,ax = axes[1])
sns.boxplot(customer_data, x="Channel", y="Frozen" ,ax = axes[2])
plt.show()
```
